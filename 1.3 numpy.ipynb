```python
org_arr = np.array([3, 1 , 9 ,5])
print(org_arr)

sort_arr1 = np.sort(org_arr)
print(sort_arr1)
print(org_arr)

sort_arr2 = org_arr.sort()
print(sort_arr2)
print(org_arr)
```

```python
arr2d = np.array([[8,12], [7,1]])

sort_arr2d_axis0 = np.sort(arr2d, axis=0)
print(sort_arr2d_axis0)

sort_arr2d_axis1 = np.sort(arr2d, axis=1)
print(sort_arr2d_axis1)
```

```python
org_arr = np.array([3, 1 , 9 ,5])
sort_indices = np.argsort(org_arr)
print(type(sort_indices))
print(sort_indices)
```

```python
name_array = np.array(['John', 'Mike', 'Sarah', 'Kate', 'Samuel'])
score_array= np.array([78, 95, 84, 98, 88])

sort_indices_asc = np.argsort(score_array)
print(sort_indices_asc)
print(name_array[sort_indices_asc])
```
