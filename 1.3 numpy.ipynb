```python
import numpy as np 

```

```python
array1 = np.array([1,2,3])
print(type(array1))
print(type(array1.shape))

array2 = np.array([[1,2,3],[4,5,6]])
print(type(array2))
print(array2.shape)

array3 = np.array([[1,2,3]])
print(type(array3))
print(array3.shape)
```

```python
print(array1.ndim)
print(array2.ndim)
print(array3.ndim)


```

```python
list1 = [1,2,3]
print(type(list1))
array1 = np.array(list1)
print(type(array1))
print(array1.dtype)





```

```python
list2 = [1, 2, 'test']
array2 = np.array(list2)
print(array2, array2.dtype)

list3 = [1, 2, 3.0]
array3 = np.array(list3)
print(array3, array3.dtype)
```

```python
array_int = np.array([1,2,3])
array_float = array_int.astype(np.float64)
print(array_float, array_float.dtype)

array_int1 = array_float.astype(np.int32)
print(array_int1, array_int1.dtype)
```

```python
sequence_array = np.arange(10)
print(sequence_array)
print(sequence_array.dtype, sequence_array.shape)
```

```python
zero_arr = np.zeros((3, 2), dtype=np.int32)
print(zero_arr)
print(zero_arr.dtype, zero_arr.shape)

one_arr = np.ones((3, 2)) 
print(one_arr)
print(one_arr.dtype, one_arr.shape)

```

```python
arr1 = np.arange(10)
print(arr1)

arr2 = arr1.reshape(2, 5)
print(arr2)

arr3 = arr1.reshape(5, 2)
print(arr3)
```

```python
arr1 = np.arange(10)
print(arr1)

arr2 = arr1.reshape(-1, 5)
print(arr2.shape)

arr3 = arr1.reshape(5, -1)
print(arr3.shape)
```

```python
arr1 = np.arange(8)
arr3d = arr1.reshape((2,2,2))
print(arr3d.tolist())

arr5 = arr3d.reshape(-1, 1)
print(arr5.tolist())
print(arr5.shape)

arr6 = arr1.reshape(-1, 1)
print(arr6.tolist())
print(arr6.shape)
```
